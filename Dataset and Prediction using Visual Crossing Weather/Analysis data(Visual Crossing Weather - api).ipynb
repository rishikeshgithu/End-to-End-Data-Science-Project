```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Bangalore Weather Data (Visual Crossing Weather).csv")

# Basic Data Exploration
print(df.head())
print(df.info())
print(df.describe())

# Visualizations
# Histograms
numeric_cols = ['Minimum Temperature', 'Maximum Temperature', 'Temperature', 'Dew Point', 'Relative Humidity', 'Wind Speed', 'Wind Gust', 'Precipitation', 'Snow Depth', 'Visibility', 'Cloud Cover', 'Sea Level Pressure']
df[numeric_cols].hist(bins=20, figsize=(15, 10))
plt.tight_layout()
plt.show()

# Time Series Plot
df['Date time'] = pd.to_datetime(df['Date time'])
plt.figure(figsize=(15, 5))
sns.lineplot(x='Date time', y='Temperature', data=df)
plt.title('Temperature Variation Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.show()

# Pair Plot
sns.pairplot(df[numeric_cols])
plt.show()

# Correlation Analysis
correlation_matrix = df[numeric_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()

# Geospatial Analysis (requires latitude and longitude columns)
# Uncomment the following lines if latitude and longitude columns are available
# plt.figure(figsize=(10, 8))
# sns.scatterplot(x='Longitude', y='Latitude', data=df, hue='Temperature', palette='coolwarm', size='Temperature', sizes=(20, 200))
# plt.title('Geospatial Distribution of Temperature')
# plt.xlabel('Longitude')
# plt.ylabel('Latitude')
# plt.legend(title='Temperature (°C)')
# plt.show()
```
