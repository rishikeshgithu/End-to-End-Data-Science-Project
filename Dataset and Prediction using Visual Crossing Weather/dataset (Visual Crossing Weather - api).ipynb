```python
import requests
import pandas as pd
from datetime import datetime, timedelta
import io

# Define the date range
start_date = datetime(2014, 1, 1)
end_date = datetime(2024, 1, 1)
interval = timedelta(days=30)

# Define the API endpoint and headers
url = "https://visual-crossing-weather.p.rapidapi.com/history"
headers = {
    "X-RapidAPI-Key": "f164b464bemsh63ea03d8724c2c1p16da60jsne602515f1682",
    "X-RapidAPI-Host": "visual-crossing-weather.p.rapidapi.com"
}

# Initialize a list to store data from each interval
all_data = []

# Loop through the date range in intervals of 30 days
current_date = start_date
while current_date < end_date:
    # Calculate the end date for this request
    next_date = min(current_date + interval, end_date)
    
    # Set the parameters for the request
    querystring = {
        "startDateTime": current_date.strftime("%Y-%m-%dT%H:%M:%S"),
        "endDateTime": next_date.strftime("%Y-%m-%dT%H:%M:%S"),
        "aggregateHours": "24",
        "location": "Bangalore,Karnataka,India",
        "unitGroup": "us",
        "dayStartTime": "8:00:00",
        "dayEndTime": "17:00:00",
        "contentType": "csv",
        "shortColumnNames": "0"
    }
    
    # Make the request
    response = requests.get(url, headers=headers, params=querystring)
    
    # Check if the request was successful
    if response.status_code == 200:
        all_data.append(response.content)
    else:
        print(f"Error: {response.status_code}")
    
    # Move to the next interval
    current_date = next_date

# Combine the data into a single DataFrame
combined_data = pd.concat([pd.read_csv(io.BytesIO(data)) for data in all_data])

# Print the combined data
print(combined_data)
```

```python
# Save the combined data to a CSV file
combined_data.to_csv('combined_weather_data.csv', index=False)
```
